```python
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)
```

```python
indep = dataset[['Age', 'EstimatedSalary', 'Gender_Male']]
dep = dataset[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
    'gamma': ['auto','scale'],
    'C':[10,100,1000,2000,3000]
}

grid = GridSearchCV(SVC(probability=True), param_grid, refit = True, verbose = 3, cv=5, n_jobs = -1, scoring='f1_weighted')
grid.fit(x_train, y_train)
```

```python
re = grid.cv_results_
print(re)
```

```python
y_predict = grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
print(cm)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_predict)
print(clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro = f1_score(y_test, y_predict, average='weighted')
print(f1_macro)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, grid.predict_proba(x_test)[:,1])
```
